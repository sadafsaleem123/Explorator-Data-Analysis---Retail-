```python
# Showing correlation using visual
fig, axes = plt.subplots(1, 1, figsize = (5,3))
sns.heatmap(df.corr())
plt.show()
```

```python
 sns.catplot(x='Category',y='Profit',data=df,col="Region")
```

```python
plt.figure(figsize=(10,10))
df.groupby(by = ['Category']).aggregate({"Profit":"sum", "Sales":"sum"}).plot.bar()

plt.show()
```

```python
plt.figure(figsize=(12,6))
df.groupby(by = ['Sub-Category']).aggregate({"Profit":"sum", "Sales":"sum"}).plot.bar()

plt.show
```

```python
plt.figure(figsize=(24,12))
df.groupby('Sub-Category')['Profit'].agg(['sum']).plot.bar()
plt.show
```

```python
sns.barplot(x=df.Region, y=df.Sales)
```

```python
sns.barplot(x=df.Region, y=df.Profit)
```

```python
sns.barplot(x=df.Region, y=df.Profit, hue=df.Category)
```

```python

dataplot = df.groupby(['State'])['Sales','Profit'].sum()
dataplot.plot.bar(figsize=(20,10))
```

```python
dataplot = df.groupby(['State'])['Profit'].sum()
dataplot.plot.bar(figsize=(20,10),color='black')
```

```python

dataplot = df.groupby(['State']).Discount.sum()
plt.xlabel('States')
plt.ylabel('Discount')
dataplot.plot.bar(figsize=(20,10),color='purple')
```
